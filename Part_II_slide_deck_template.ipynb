## Investigation Overview


The primary objectives of this presentation are to demonstrate the dominant industries housing the most unicorn companies, offer insights into the typical timeline for startups to attain unicorn status, delineate the countries with the most unicorns and the cities emerging as industry hubs, and finally, shed light on influential venture capital firms consistently backing unicorn companies.


## Dataset Overview

The data we are using is 'Unicorn Companies' dataset, contains private companies the took the name unicorn, that means the company got a valuation over a 1 billion US dollar, the data has 1074 entries, and 10 columns including the name of the company, the valuation, funding, county of origin, investors, the year founded and the year become unicorn.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('Unicorn+Companies/df_clean.csv')
```

## Top 10 Industries Having Unicorn Companies

Technology-related industries, such as Fintech, Software, E-commerce, and Artificial Intelligence, have the highest concentration of unicorn companies.
Emerging sectors like HealthTech and Supply chain are also seeing a growing number of unicorn startups.
Traditional industries like Real Estate are less represented among unicorn companies.


```python
# Q: Which industry has the most interest for unicorn companies?

plt.figure(figsize=[14, 6])
sb.countplot(data= df, y='Industry', order=df.Industry.value_counts().iloc[:10].index, color= sb.color_palette()[5]);
plt.title('Top 10 Industries have Unicorn Companies')
plt.xlabel('Count of Unicorn Companies');
```

## Years Needed for a Company to Hit 1B Valuation

The time it takes for a company to become a unicorn varies significantly by industry and market conditions.
On average, it takes around 7 years from founding to achieve unicorn status, but some startups achieve this status much faster or slower due to other factors not included in our data.

```python
# Q: How long does it usually take for a company to become a unicorn?

plt.figure(figsize=[14, 6])
plt.hist(data= df, x= 'duration_to_unicorn', bins= 100, color= sb.color_palette()[5]);
plt.xlim(0, 30);
plt.xlabel('Duration (Year)')
plt.ylabel('Count of Unicorn Companies')
plt.title('Number of years took the companies to reach 1B valuation');
```

## The Largest Unicorn Ecosystem
### The Top 10 Cities have Unicorn Companies

The United States, particularly Silicon Valley, remains the global leader in unicorn companies, with a high concentration in cities like San Francisco and New York.
China, especially cities like Beijing and Shanghai, has seen a significant rise in unicorn startups, making it the second-largest unicorn ecosystem.
Other countries in Asia, such as India and Singapore, are also emerging as unicorn hubs.
European countries like the United Kingdom, Germany, and France have notable unicorn ecosystems, with London and Berlin as key cities.

```python
# Q: Which City has the most unicorns?

plt.figure(figsize=[14, 6])
sb.countplot(data= df, y='City', order=df.City.value_counts().iloc[:10].index, color= sb.color_palette()[5]);
plt.title('Top 10 Cities have Unicorn Companies')
plt.xlabel('Count of Unicorn Companies');
```

## Top 10 Investors Based on The Number of Their Investments in Unicorn Companies

Prominent venture capital firms like Sequoia Capital, Andreessen Horowitz, and Accel have consistently been major backers of unicorn companies.

```python
investors_df = pd.DataFrame(columns=['investor', 'count'])

for i in ['investor1','investor2', 'investor3', 'investor4']:
    dff = df[i].value_counts().rename_axis('investor').reset_index(name= 'count')
    investors_df = pd.concat([investors_df,dff], ignore_index= True)

investors_df['investor'] = investors_df['investor'].str.strip()    
investors_df = investors_df.astype({'count': 'int32'})
```

```python
#which are the most 10 investing in companies that hit the 1B?

plt.figure(figsize=[14, 6])
sb.barplot(data= investors_df.head(10), x= 'count', y= 'investor', color=sb.color_palette()[5])
plt.title('Top 10 investors by number of investments in unicorns');
plt.xlabel('Count of Unicorn Companies');
```
